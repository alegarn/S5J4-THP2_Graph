# 2.7. Graphe 7

Nous souhaitons un graphe qui donne l'√©volution du CA pour chaque cat√©gorie (en fonction du temps).

```python
file['day'] = file['date'].str[:11]
file['month'] = file['date'].str[:7]
per_day = file.groupby('day', as_index=False)
per_month = file.groupby('month', as_index=False)
file['price'].astype(float)

monthly_turnover = per_month['price'].sum().add_suffix('_sum')
daily_turnover = per_day['price'].sum().add_suffix('_sum')

file_cat_0 = file.loc[file["categ"] == 0].sort_values('price', ascending = False)
file_cat_1 = file.loc[file["categ"] == 1].sort_values('price', ascending = False)
file_cat_2 = file.loc[file["categ"] == 2].sort_values('price', ascending = False)

per_month_0 = file_cat_0.groupby('month', as_index=False)
per_month_1 = file_cat_1.groupby('month', as_index=False)
per_month_2 = file_cat_2.groupby('month', as_index=False)

monthly_turnover_cat_0 = per_month_0['price'].sum().add_suffix('_sum_cat_0')
monthly_turnover_cat_1 = per_month_1['price'].sum().add_suffix('_sum_cat_1')
monthly_turnover_cat_2 = per_month_2['price'].sum().add_suffix('_sum_cat_2')

monthly_turnover['price_sum_cat_0'] = monthly_turnover_cat_0['price_sum_cat_0']
monthly_turnover['price_sum_cat_1'] = monthly_turnover_cat_1['price_sum_cat_1']
monthly_turnover['price_sum_cat_2'] = monthly_turnover_cat_2['price_sum_cat_2']

monthly_turnover.set_index('month_sum')
month = monthly_turnover['month_sum'].to_numpy()
```

```python

price_sum_cat_0 = monthly_turnover['price_sum_cat_0'].to_numpy()
price_sum_cat_1 = monthly_turnover['price_sum_cat_1'].to_numpy()
price_sum_cat_2 = monthly_turnover['price_sum_cat_2'].to_numpy()

plt.plot(month, price_sum_cat_0, label = "profit_cat_0")
plt.plot(month, price_sum_cat_1, label = "profit_cat_1")
plt.plot(month, price_sum_cat_2, label = "profit_cat_2")

x = np.arange(len(month))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*0.8, price_sum_cat_0, width, label='profit_cat_0')
rects2 = ax.bar(x + width/5, price_sum_cat_1, width, label='profit_cat_1')
rects3 = ax.bar(x + width*1.2, price_sum_cat_2, width, label='profit_cat_2')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Cat_profits')
ax.set_title('√©volution du CA pour chaque cat√©gorie')
ax.set_xticks(x, month.any())
ax.legend()


fig.tight_layout()

#plt.bar(monthly_turnover['month_sum'], monthly_turnover['price_sum_cat_0'], align='center', alpha=0.5)
```
